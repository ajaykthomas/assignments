# pca

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(data)
x_pca.shape
```

```python
#From 15 columns, we came down to 2 columns
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x_pca[:, 0], x_pca[:,1], lw = 2)
plt.xlabel('First Principle Component')
plt.ylabel('Second Principle Component')
plt.title('Principle Component Analysis')
plt.show()
```

# clustering model

# k mean clustering

```python
# find the optimal number of cluster in k mean clustering
from sklearn.cluster import KMeans
# next we have to cretae instant
```

```python
inertia=[]
```

```python
# for loop is used by the  optimal number of cluster based on the inertia value

for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)
    print(kmeans.inertia_)
  
```

```python
plt.plot(range(1,11),inertia)
plt.title("the elbow method")
plt.xlabel("number of cluster")
plt.ylabel("inertia")
plt.show()
```

```python
# the optimal number of cluster 3
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = pd.Series(kmeans.fit_predict(data))
y_kmeans
```

```python
#number of datapoint in cluster
y_kmeans.value_counts()
```

```python
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids) #centroids of principle components
```

```python
centroids
```

```python
centroids_pca
```

```python
pca_df = pd.DataFrame(x_pca, columns = ['PCA 1', 'PCA 2'])
pca_df['Cluster'] = y_kmeans
pca_df
```

```python
# visulizing
plt.figure(figsize=(12,8))
sns.scatterplot(x = 'PCA 1', y = 'PCA 2', hue = 'Cluster', data = pca_df, palette = ['red', 'green', 'orange'])
plt.scatter(x = centroids_pca[:, 0], y = centroids_pca[:, 1],marker = '.', color = 'blue', s=300)
plt.title('Clusters')
plt.show()
```

# hieraarchial clustering

```python
# use the dendogram to find opitmum number of cluster
# import the libary
```

```python
import scipy.cluster.hierarchy as sch
```

```python
#create instance of dendorgarm with x_pca
dendrogram=sch.dendrogram(sch.linkage(x_pca, method='ward'))
plt.title('dendrogram')
plt.xlabel('customers')
plt.ylabel('euclidean distyance')
plt.show()
```

```python
#optimal number of cluster is 3
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
```

```python
y_hc=hc.fit_predict(x_pca)
```

```python
y_hc
```

```python
plt.scatter(x_pca[y_hc==0,0],x_pca[y_hc==0,1],s=100, c='red',label="cluster 1")
plt.scatter(x_pca[y_hc==1,0],x_pca[y_hc==1,1],s=100, c='blue',label="cluster 2")
plt.scatter(x_pca[y_hc==2,0],x_pca[y_hc==2,1],s=100, c='green',label="cluster 3")
plt.title("cluster of cluster")
plt.xlabel("pca1")
plt.ylabel("pca2")
plt.legend()
plt.show()
```
